```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\DELL\Desktop\train.csv file.CSV")
df
```

```python
a=df.isnull().sum()
a
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Age"]
```

```python
df["Embarked"]= df["Embarked"].fillna(df["Embarked"].mode()[0])
df["Embarked"]
```

```python
df = df.drop(columns=["Name","Ticket","Cabin"])
```

```python
display(df)
```

```python
plt.figure()
plt.hist(df["Age"],bins=6)
plt.title("Age Distribution Histogram")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
df["Age"].plot(kind="kde")
plt.title("Age Distribution KDE")
plt.show()
```

```python
plt.figure()
df["Sex"].value_counts().plot(kind="bar")
plt.title("Passenger Count by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
plt.boxplot(df["Fare"])
plt.title("Fare Distribution")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure()
pd.crosstab(df["Sex"],df["Survived"]).plot(kind="bar")
plt.title("Survial Count by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
df.boxplot(column="Fare", by="Pclass")
plt.title("Fare by Passenger Class")
plt.suptitle("")
plt.xlabel("Pclass")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure()
plt.scatter(df["Age"], df["Fare"], c=df["Survived"])
plt.title("Age vs Fare (Colored by Survival)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
